This is a project on Car Price prediction. A chinese company would like to enter the US market, first the company would like to know the factors that affect the prices of Cars in America. Therefore data was taken from 205 cars, varying models each with varying specifications fuel type, number of doors, et cetera, with different prices. Six models were build based on the data and the accuracy of each measured using the R squared, the mean absolute error(MAE) and the Mean Absolute Percentage Error (MAPE).


```python
egression, LassoCV, ElasticNetCV
from sklearn.model_selection 
import GridSearchCV
import pandas as pd   #data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np    #linear algebra
import matplotlib.pyplot as plt #Plotting and Visualizing data
import seaborn as sns    #Visualizing dataset structure
import os    # Allows python to interface with the underlying operating system
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge, LinearR
```

```python
data=pd.read_csv('CarPriceProject.csv')
```

```python
#To check for any missing values in the data
data.isnull().sum()
```

```python
#Separating dependedent & Independent Variables 
X=data.iloc[:,1:-1]
y=data['price']
X=X.drop(['CarName'],axis=1)
```

```python
X.columns
```

```python
#To convert categorical variables to numeric variable(indicator variables)
X_trans = pd.get_dummies(X)
X_trans.head()
```

```python
#To split the data, 80% training set and 20% test set
X_new,X_test,y_new,y_test=train_test_split(X_trans,y,test_size=0.2, random_state=0)
X_new.head()
X_test.shape
```

```python
#To check for missing values
X_new.isnull().sum()
```

```python
cat=X_new.dtypes=='object'
cat_f=list(cat[cat].index)
print(cat_f)
```

```python
X_test.shape
```

```python
model = LinearRegression()
model.fit(X_new, y_new)

acc_train, acc_test = model.score(X_new, y_new), model.score(X_test, y_test)
mae_train = mean_absolute_error(y_new, model.predict(X_new))
mae_test = mean_absolute_error(y_test, model.predict(X_test))
mape_train = (np.abs(y_new - model.predict(X_new)) / y_new).mean()
mape_test = (np.abs(y_test - model.predict(X_test)) / y_test).mean()

print('\033[1mTraining\033[0m\nR^2: {0:.4f}'.format(acc_train))
print('Mean Absolute Error: ${0:,.2f}'.format(mae_train))
print('Mean Absolute Percent Error: {0:.2%}'.format(mape_train))

print('\n\033[1mTesting\033[0m\nR^2: {0:.4f}'.format(acc_test))
print('Mean Absolute Error: ${0:,.2f}'.format(mae_test))
print('Mean Absolute Percent Error: {0:.2%}'.format(mape_test))
```

```python
model = GridSearchCV(Ridge(), param_grid={'alpha':np.arange(0.1, 2, 0.1)})
model.fit(X_new, y_new)
print('Best alpha: {0:.1f}\n'.format(model.best_estimator_.alpha))

acc_train, acc_test = model.score(X_new, y_new), model.score(X_test, y_test)
mae_train = mean_absolute_error(y_new, model.predict(X_new))
mae_test = mean_absolute_error(y_test, model.predict(X_test))
mape_train = (np.abs(y_new - model.predict(X_new)) / y_new).mean()
mape_test = (np.abs(y_test - model.predict(X_test)) / y_test).mean()

print('\033[1mTraining\033[0m\nR^2: {0:.4f}'.format(acc_train))
print('Mean Absolute Error: ${0:,.2f}'.format(mae_train))
print('Mean Absolute Percent Error: {0:.2%}'.format(mape_train))

print('\n\033[1mTesting\033[0m\nR^2: {0:.4f}'.format(acc_test))
print('Mean Absolute Error: ${0:,.2f}'.format(mae_test))
print('Mean Absolute Percent Error: {0:.2%}'.format(mape_test))
```

```python
model = LassoCV()
model.fit(X_new, y_new)

acc_train, acc_test = model.score(X_new, y_new), model.score(X_test, y_test)
mae_train = mean_absolute_error(y_new, model.predict(X_new))
mae_test = mean_absolute_error(y_test, model.predict(X_test))
mape_train = (np.abs(y_new - model.predict(X_new)) / y_new).mean()
mape_test = (np.abs(y_test - model.predict(X_test)) / y_test).mean()

print('\033[1mTraining\033[0m\nR^2: {0:.4f}'.format(acc_train))
print('Mean Absolute Error: ${0:,.2f}'.format(mae_train))
print('Mean Absolute Percent Error: {0:.2%}'.format(mape_train))

print('\n\033[1mTesting\033[0m\nR^2: {0:.4f}'.format(acc_test))
print('Mean Absolute Error: ${0:,.2f}'.format(mae_test))
print('Mean Absolute Percent Error: {0:.2%}'.format(mape_test))
```

```python
model = ElasticNetCV()
model.fit(X_new, y_new)

acc_train, acc_test = model.score(X_new, y_new), model.score(X_test, y_test)
mae_train = mean_absolute_error(y_new, model.predict(X_new))
mae_test = mean_absolute_error(y_test, model.predict(X_test))
mape_train = (np.abs(y_new - model.predict(X_new)) / y_new).mean()
mape_test = (np.abs(y_test - model.predict(X_test)) / y_test).mean()

print('\033[1mTraining\033[0m\nR^2: {0:.4f}'.format(acc_train))
print('Mean Absolute Error: ${0:,.2f}'.format(mae_train))
print('Mean Absolute Percent Error: {0:.2%}'.format(mape_train))

print('\n\033[1mTesting\033[0m\nR^2: {0:.4f}'.format(acc_test))
print('Mean Absolute Error: ${0:,.2f}'.format(mae_test))
print('Mean Absolute Percent Error: {0:.2%}'.format(mape_test))
```
